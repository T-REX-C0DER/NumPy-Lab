```python
import numpy as np 
```

# Vector Operations 

```python
# Vector addition
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 + arr2
print("Vector addition result:", result)
```

```python
# Vector subtraction
result = arr1 - arr2
print("Vector subtraction result:", result) 
```

```python
# Scalar multiplication
scalar = 2
result = arr1 * scalar
print("Scalar multiplication result:", result)
```

```python
# Dot product
dot_product = np.dot(arr1, arr2)
print("Dot product result:", dot_product)
```

```python
# Vector magnitude
magnitude = np.linalg.norm(arr1)
print("Vector magnitude:", magnitude)
```

```python
# unit vector
unit_vector = arr1 / magnitude
print("Unit vector:", unit_vector)
```

```python
# Cross product (for 3D vectors)
cross_product = np.cross(arr1, arr2)
print("Cross product result:", cross_product)
```

```python
# Angle between two vectors
cos_angle = np.dot(arr1, arr2) / (np.linalg.norm(arr1) * np.linalg.norm(arr2))
angle = np.arccos(cos_angle)
print("Angle between vectors (in radians):", angle)
print("Angle between vectors (in degrees):", np.degrees(angle))
```

```python
# Projection of arr1 onto arr2
projection = (np.dot(arr1, arr2) / np.dot(arr2, arr2)) * arr2
print("Projection of arr1 onto arr2:", projection)
```

```python
# Distance between two vectors
distance = np.linalg.norm(arr1 - arr2)
print("Distance between arr1 and arr2:", distance)
```
