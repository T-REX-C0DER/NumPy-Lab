#  Declaration Of Arrays In NumPy

```python
import numpy as np

a = np.array([23,56,43,89,12]) # Simple 1D array declaration in the numpy
print("Array a is :", a)
```

```python
import numpy as np
# Diff Between Lists & Arrays 
list = [1,2,3,4,5]
print(f"The multiplication using lists will print this : {list*2}") # This will print the list two times

array = np.array([1,2,3,4,5])
print(f"The multiplication using arrays will print this : {array*2}") # This will multiply each element by 2
```

# Creating Arrays With Default Values

```python
import numpy as np

# Array Of Zeros 
a1 = np.zeros(4) # 1D array 
a2 = np.zeros((3,3)) # 2D array
print(a1)
print(a2)

# Array Of Ones
b1 = np.ones(3) # 1D array
b2 = np.ones((5,6))
print(b1)
print(b2)

# Array Of full
c2 = np.full((3,4),7) # 2D array it wil print the given fixed values 
print(c2)

# Array Of random values
d1 = np.random.random(5)
d2 = np.random.random((3,4))
print(d1)
print(d2)

# Array Of range
e1 = np.arange(0,10,1) # (start , stop , step)
print(e1)

# Array Of linspace
f1 = np.linspace(0,10,5) # (start , stop , number of values)
print(f1)

# Array Of eye
g1 = np.eye(4) # Identity matrix of size 4x4    
print(g1)
```

# Array Properties 

```python
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]]) # original array
print("Original array:", arr) 

print(arr.ndim)  # number of dimensions

print(arr.shape)  # shape of the array

print(arr.size)  # total number of elements

print(arr.dtype)  # data type of the elements

print(arr.itemsize)  # size of each element in bytes

print(arr.nbytes)  # total size of the array in bytes

print(arr.T)  # transpose of the array

print(arr.flatten())  # flatten the array to 1D

print(arr.reshape(3, 2))  # reshape the array to 3x2

print(arr.sum())  # sum of all elements

print(arr.min())  # minimum element

print(arr.max())  # maximum element

print(arr.mean())  # mean of the elements

print(arr.std())  # standard deviation of the elements

print(arr.var())  # variance of the elements 

print(np.unique(arr))  # unique elements in the array
```
